```python
for col in df.columns:
    print(f"Einzigartige Werte in {col}:")
    print(df[col].unique())
    print("-" * 50)
```

```python
df.replace("?", np.nan, inplace=True)
display(df.isnull().sum())
display(df.isna().sum())
```

```python
# Prozentuale Verteilung der fehlenden Werte
print(df["workclass"].value_counts(normalize=True))
print(df["occupation"].value_counts(normalize=True))
print(df["native.country"].value_counts(normalize=True))
```

Entscheidung für workclass: 
Private macht 73,86 % aus, das ist eine sehr dominante Kategorie.
Empfehlung: Mode (Private) auffüllen, da es höchstwahrscheinlich zutrifft.

Entscheidung für occupation:
Die Häufigkeiten sind relativ gleichmäßig verteilt. Keine einzelne Kategorie dominiert.
Empfehlung: "Unknown" verwenden, da es sonst zu Verzerrungen kommt.

Entscheidung für native.country:
United-States macht 91,22 % aus, extrem dominant.
Empfehlung: Mode (United-States) ist hier sinnvoll.

```python
df["workclass"] = df["workclass"].fillna("Private")
df["occupation"] = df["occupation"].fillna("Unknown")
df["native.country"] = df["native.country"].fillna("United-States")
```

```python
display(df.isnull().sum())
display(df.isna().sum())
```
